# üßπ Data Cleaning & Preprocessing

<span style="color: #FFD700; font-size: 20px;">
  üß© Missing Data Handling
</span>

```python
# Check for missing data in each variable
print("üîé Number of missing values per variable:\n")
print(df.isnull().sum().sort_values(ascending=False))

print("="*100)

# How many observations have at least one missing value?
missing_obs = df.isnull().any(axis=1).sum()
missing_ratio = (missing_obs / df.shape[0]) * 100

print(f"‚ö†Ô∏è Observations with at least one missing value: {missing_obs} rows")
print(f"üìâ This represents approximately {missing_ratio:.2f}% of the dataset")

# Drop rows with missing data
df.dropna(inplace=True)
print("‚úÖ Missing data rows have been dropped.")
```

<span style="color: #FFD700; font-size: 20px;">
  üóÇÔ∏è Duplicate Rows Check & Removal
</span>

```python
# üîç Check and remove duplicate rows

# Count duplicate rows
duplicate_count = df.duplicated().sum()
print(f"üóÇÔ∏è Number of duplicate rows: {duplicate_count}")

# Remove duplicates if any exist
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("‚úÖ Duplicate rows have been removed.")
else:
    print("üéâ No duplicate rows found!")
```

<span style="color: #FFD700; font-size: 20px;">
  üîÑ Reset DataFrame Index
</span>

```python
# Reset index and discard the old one
df.reset_index(drop=True, inplace=True)
```

<span style="color: #FFD700; font-size: 20px;">
  üìã Column Data Types Overview
</span>

```python
# Check data types of all columns
df.dtypes
```

```python
# Change data type of columns only for better visualization

df = df.astype({
    "votes": "int64",
    "budget": "int64",
    "gross": "int64",
    "runtime": "int64"
})
```

```python
df.head(3)
```

<span style="color: #FFD700; font-size: 15px; font-weight: bold; display: block; margin-top: 5px;">
  ‚ö†Ô∏è We focus on the 'rating' variable: combining 'Unrated' and 'Not Rated' into a single category
</span>


```python
df["rating"].value_counts()
```

```python
# Combine equivalent rating labels
df["rating"] = df["rating"].replace({"Unrated": "Not Rated"})
```

# üìäüîó Correlation Analysis

<span style="color: #FFD700; font-size: 20px; display: block; margin-top: 15x;">
  üî¢ Correlation Analysis - Numeric Features Only
</span>

```python
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
# Visualize all pairwise correlations in a stacked format
sorted_pairs = correlation_matrix.stack()
sorted_pairs
```

```python
# Filter and view only strong correlations (absolute value > 0.5, exclude self-correlation)
strong_pairs = sorted_pairs[(abs(sorted_pairs) > 0.5) & (sorted_pairs < 1.0)].drop_duplicates()
strong_pairs
```

```python
#Visualizing the correlation matrix to highlight relationships between numerical features.

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numerical Features", fontsize=16, fontweight='bold', pad=14)
plt.show()
```

# üîç Asking Questions and Finding Answers

<span style="color: #FFD700; font-size: 20px; display: block; margin-top: 10px;">
  ‚≠ê What are the 5 most talked-about movies of all time? 
</span>

```python
df.sort_values(by=['votes'], ascending=False).reset_index(drop=True)[["name"]].head(10)
```

<span style="color: #FFD700; font-size: 20px; display: block; margin-top: 10px;">
  üèÜ Top Earning Companies by Total Gross 
</span>

```python
# Total gross per company
total_gross_per_company = df.groupby("company")[["gross"]].sum()

# Sorted from Highest to Lowest
total_gross_sorted = total_gross_per_company.sort_values(by="gross", ascending=False)
total_gross_sorted.head(10)  
```

<span style="color: #FFD700; font-size: 20px; display: block; margin-top: 10px;">
  üí∞ Most Profitable Companies by Average Movie Revenue 
</span>

```python
# Calculate the average gross revenue per company and sort from highest to lowest
avg_gross_per_company = (
    df.groupby("company")["gross"]
    .mean()
    .sort_values(ascending=False)
)

# Convert the Series to a DataFrame and rename the column for clarity
avg_gross_per_company_df = avg_gross_per_company.reset_index()
avg_gross_per_company_df.columns = ["Company", "Average Gross Revenue"]

# Display the top 10 companies with the highest average gross revenue
avg_gross_per_company_df.head(10)
```
